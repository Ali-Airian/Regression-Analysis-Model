```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\ideal.csv')
testdata=pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\test.csv')
traindata=pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\train.csv')
print(data.shape)
print(testdata.shape)
print(traindata.shape)
```

```python
def leastsquarefunc(yf):
    yt='y'+str(yf)
    print("ideal data"+' '+ yt)
    X = data['x'].values
    Y = data['y'+str(yf)].values
    mean_x = np.mean(X)
    mean_y = np.mean(Y)
    # Total number of values
    n = len(X)
    # Using the formula to calculate 'm' and 'c'
    numer = 0
    denom = 0
    for i in range(n):
        numer += (X[i] - mean_x) * (Y[i] - mean_y)
        denom += (X[i] - mean_x) ** 2
        m = numer / denom
        c = mean_y - (m * mean_x)
    # Printing coefficients
    print("Coefficients")
    print(m,' , ' ,c)
    
    xn = len(X)
    rmse = 0
    for i in range(xn):
        y_pred = c + m * X[i]
        rmse += (Y[i] - y_pred) ** 2
    rmse = np.sqrt(rmse/n)
    print("RMSE")
    print(rmse)
    
    ss_tot = 0
    ss_res = 0
    for i in range(xn):
        y_pred = c + m * X[i]
        ss_tot += (Y[i] - mean_y) ** 2
        ss_res += (Y[i] - y_pred) ** 2
     
    r2 = 1 - (ss_res/ss_tot)
    print("R2 Score")
    print(r2)
    
    max_x = np.max(X) + 100
    min_x = np.min(X) - 100
    # Calculating line values x and y
    x = np.linspace(min_x, max_x, 1000)
    y = c + m * x
    # Ploting Line
    plt.plot(x, y, color='#58b970', label='Regression Line')
    # Ploting Scatter Points
    plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
    plt.xlabel('x')
    plt.ylabel(yt)
    plt.legend()
    plt.show()
    
    print('***************#####*************************') 

    
```

```python
for index in range(data.shape[1]):
    #print('Column Number : ', index)
    if(index!=0):  
         i=str(index)
         leastsquarefunc(i)
```
