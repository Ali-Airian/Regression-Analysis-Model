```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\ideal.csv')
testdata=pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\test.csv')
traindata=pd.read_csv(r'C:\Users\Muhammad Luqman\DLMDSPWP01_Python Assigment\train.csv')
```

```python
traindata['intercept'] = 1 
```

```python
def RegFuncTrain(yf):
    yt='y'+str(yf)
    print("train data"+' '+ yt)
    
    lm = sm.OLS(traindata[yt], traindata[['intercept', 'x']]).fit()
    print(lm.summary())
    print('***************#####*************************')
```

```python
vartrainitterate=traindata.shape[1] -1
for index in range(vartrainitterate):
    #print('Column Number : ', index)
    if(index!=0):  
         i=str(index)   
         RegFuncTrain(i)
```
